```python
import math as ma
import numpy as np
import matplotlib.pyplot as plt
import shapely
from shapely.geometry import LineString, Point

T = np.linspace(1,float(10E9))
epp = 1.08*10**(-7)*150*(0.7)**2*(T/10**6)**4
ecno = 8.24*10**(-26)*150*0.0081*.7*(T/10**6)**(19.9)
#Title and axis 
plt.title("Star Fusion", fontsize=14)
plt.xlabel('log [T/K]', fontsize=14)
plt.ylabel('log(epsilon [erg s^-1 g^-1]', fontsize=14) 

#Plots of the functions
plt.plot(np.log10(T),np.log10(epp),'maroon',label = "e_pp")
plt.plot(np.log10(T), np.log10(ecno), color='green',label = "e_cno")

plt.scatter(7.26,0, color = 'b', lw = 10, marker = '+', label = "Equal Temperature")
plt.scatter(7.2, np.log10(0.482),color = 'black', lw = 10, marker = '+', label = "Suns Central Temp")


#setting the x and y range per the assignment 
plt.xlim(7,8)
plt.ylim(-10, 20)
plt.legend(loc=0)




```
